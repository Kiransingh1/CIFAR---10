```python
import numpy as np
import cifar10
import matplotlib.pyplot as plt
from sklearn import ensemble 
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
from sklearn import svm
from sklearn.model_selection import GridSearchCV
```

```python
cifar10.data_path = "data/CIFAR-10/"
```

```python
cifar10.maybe_download_and_extract()
```

```python
class_names = cifar10.load_class_names()
class_names
```

```python
images_train, cls_train, labels_train = cifar10.load_training_data()
images_test, cls_test, labels_test = cifar10.load_test_data()
```

```python
images_train.shape
```

```python
images_train = images_train.reshape(50000,-1)
images_test = images_test.reshape(10000, -1)
```

```python
pca = PCA()
pca.fit(images_train)
```

```python
k = 0
total = 0
while total < 0.99:
    total += pca.explained_variance_ratio_[k]
    k += 1
k
```

```python
pca_1 = PCA(n_components = k, whiten= True)
images_train = pca_1.fit_transform(images_train)
images_test= pca_1.transform(images_test)
```

```python
lr_clf = LogisticRegression(C=0.01)
lr_clf.fit(images_train, cls_train)
y_pred = lr_clf.predict(images_test)
accuracy_score(cls_test, y_pred)
```

```python
rf_clf = ensemble.RandomForestClassifier(max_depth= 10)
rf_clf.fit(images_train, cls_train)
y_pred = rf_clf.predict(images_test)
accuracy_score(cls_test, y_pred)
```
